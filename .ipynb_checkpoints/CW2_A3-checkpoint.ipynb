# Imports 

```python
from nltk.corpus import brown
import nltk
import string
import pandas as pd
from nltk.corpus import wordnet as wn
from collections import Counter
import time
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sb
stopwords = list(set(nltk.corpus.stopwords.words("English")))
punctuation = list(string.punctuation)
stop_with_punctuation = stopwords+punctuation
stop_with_punctuation.append("--")
stop_with_punctuation.append("``")
stop_with_punctuation.append("''")
```

# Activity 3
## 3.1 Sub-activity: Loading the Brown corpus data from NLTK
### Task 1

```python
# Get the categories of corpus. 
brown_categories = brown.categories()
# Sort in alphabetical order.
brown_categories.sort()
print(brown_categories)
```

### Task2

```python
# Initialise dataframe
start_time = time.time()
# (a). Add five columns to the dataframe. 
brown_corpus = pd.DataFrame(columns = ["category_name", "count_words", "count_sentences", "count_paragraphs", "common_words"])
print("start")

# Prepare the data
data = [[category, len(brown.words(categories=category)), 
         len(brown.sents(categories=category)), len(brown.tagged_paras(categories=category)),
         [item[0] for item in nltk.FreqDist(word for word in brown.words(categories=category)).most_common(10)]] 
        for category in brown_categories
       ]

# (b). Fill the DataFrame with data. 
for datum in data:
    brown_corpus = brown_corpus.append({"category_name": datum[0], "count_words": datum[1], "count_sentences":datum[2], "count_paragraphs":datum[3],"common_words":datum[4]}, ignore_index = True)
time_consumption = time.time() - start_time
print("end")
print("{}s passed!".format(time_consumption))

# (c). Add a new column to the DataFrame
brown_corpus["common_words_after_preprocessing"] = brown_corpus.apply(lambda row: [item[0] for item in nltk.FreqDist(word for word in brown.words(categories=row.category_name) if word.lower() not in stop_with_punctuation).most_common(10)], axis=1)

# (d). print the first ten rows of the final dataframe
brown_corpus[:10]
```

## 3.2 Sub-activity: Applying NLP operations on the corpus
### 3.2.1 Stemming
#### Task 3

```python
news_brown = brown.words(categories="news")
# (a) First, remove stopwords and punctuation from the list of words of category news
news_words = [word for word in news_brown if word.lower() not in stop_with_punctuation]
news_freq_dist = nltk.FreqDist(news_words)
unique_words = news_freq_dist.keys()

# (b) Print the number of unique words in news words.
print("Before using PorterStemmer to stem, the size of unique is {}".format(len(unique_words)))

# (c). Apply the PorterStemmer  
porter = nltk.PorterStemmer()
news_words_after_stemmed = [porter.stem(word) for word in news_words]
news_freq_dist = nltk.FreqDist(news_words_after_stemmed)
unique_words = news_freq_dist.keys()
print("After using PorterStemmer to stem, the size of unique is {}".format(len(unique_words)))
```

#### Task 4 

```python
fiction_raw_words = brown.words(categories="fiction")

# (a) First, remove stopwords and punctuation from the list of words of category fiction
fiction_words = [word for word in fiction_raw_words if word.lower() not in stop_with_punctuation]
fiction_freq_dist = nltk.FreqDist(fiction_words)
unique_words = fiction_freq_dist.keys()

# (b) Print the number of unique words in news words.
print("Before using SnowballStemmer to stem, the size of unique is {}".format(len(unique_words)))

# (c). Apply the SnowballStemmer  
porter = nltk.SnowballStemmer("english")
fiction_words_after_stemmed = [porter.stem(word) for word in fiction_words]
fiction_freq_dist = nltk.FreqDist(fiction_words_after_stemmed)
unique_words = fiction_freq_dist.keys()
print("After using SnowballStemmer to stem, the size of unique is {}".format(len(unique_words)))
```

### 3.2.2 Lemmatization
#### Task 5

```python
government_raw_words = brown.words(categories="government")

# (a) First, remove stopwords and punctuation from the list of words of category government
government_words = [word.lower() for word in government_raw_words if word.lower() not in stop_with_punctuation]
government_freq_dist = nltk.FreqDist(government_words)
unique_words = government_freq_dist.keys()

# (b) Print the number of unique words in news words.
print("Before using WordNetLemmatizer to lemmatize, the size of unique is {}".format(len(unique_words)))

# (c). Apply the WordNetLemmatizer   
wnl = nltk.stem.WordNetLemmatizer()
government_words_after_lemmatized = [wnl.lemmatize(word) for word in government_words]
government_freq_dist = nltk.FreqDist(government_words_after_lemmatized)
unique_words = government_freq_dist.keys()
print("After using WordNetLemmatizer to lemmatize, the size of unique is {}".format(len(unique_words)))
```

### 3.2.3 Finding Synonyms and Anatonyms
#### Task 6

```python
brown_corpus["synonyms"] = ""
brown_corpus["antonyms"] = ""
def get_synonyms(common_words:list) -> list:
    """
        Takes as its input a list of words and returns a list of all their synonyms 
        :param common_words: a list of common words after removing stopwords and punctuations
        :return: list of common_words' synonyms
    """
    all_common_words = []
    for word in common_words:
        for synset in wn.synsets(word):
            for lemma in synset.lemmas():
                all_common_words.append(lemma.name())
    return list(set(all_common_words))

def get_antonyms(common_words:list) -> list:
    """
        Takes as its input a list of words and returns a list of all their antonyms 
        :param common_words: a list of common words after removing stopwords and punctuations
        :return: list of common_words' antonyms
    """
    all_common_words_antonyms = []
    for word in common_words:
        for synset in wn.synsets(word):
            for lemma in synset.lemmas():
                if lemma.antonyms():
                    all_common_words_antonyms.append([item.name() for item in lemma.antonyms()][0])
    return list(set(all_common_words_antonyms))

brown_corpus["synonyms"] = brown_corpus.apply(lambda row: get_synonyms(row.common_words_after_preprocessing), axis=1)
brown_corpus["antonyms"] = brown_corpus.apply(lambda row: get_antonyms(row.common_words_after_preprocessing), axis=1)

# (d) Print the first ten rows of the dataframe.
brown_corpus[:10]
```

### 3.2.4 Bigrams and trigrams
#### Task 7

```python
# Select last 500 paragraphs 
last_500_paragraph = brown.tagged_paras()[-500:]
assert len(last_500_paragraph) == 500
print("The list length is {}".format(len(last_500_paragraph)))
```

#### Task 8

```python
def get_bigrams_frequency(paragraph_list: list) -> dict:
    """
        Takes as input a list of paragraphs and returns a dictionary where key is the bigram and the value is the frequency
        :param paragraph_list 
        :return a dictionary where key is the bigram and the value is the frequency
    """
    # Remove stopwords and puncuation tokens
    last_500_paragraph_sentence = [sentence for paragraph in last_500_paragraph for sentence in paragraph]
    last_500_paragraph_words = [word[0] for sentence in last_500_paragraph_sentence for word in sentence]
    words_after_screening = [word for word in last_500_paragraph_words if word.lower() not in stop_with_punctuation]
    bgs = nltk.bigrams(words_after_screening)
    fdist = nltk.FreqDist(bgs)
    return dict(fdist)
```

#### Task 9

```python
brown_bigrams = get_bigrams_frequency(last_500_paragraph)
print("Save brown_bigrams as a {}".format(type(brown_bigrams)))
print("The size of the bigram is {}.".format(len(brown_bigrams)))
```

#### Task 10

```python
top_15_bigrams = sorted(brown_bigrams, key=brown_bigrams.get, reverse=True)[:15]
print("The top 15 occuring bigrams are as follows.")
top_15_bigrams
```

## 3.3.2 Heatmap
### Task 14

```python
def get_bigrams_frequency_by_word_list(word_list):
    """
        Takes as input a list of words and returns a dictionary where key is the bigram and the value is the frequency
        :param word_list 
        :return a dictionary where key is the bigram and the value is the frequency
    """
    # Remove stopwords and puncuation tokens
    words_after_screening = [word for word in word_list if word.lower() not in stop_with_punctuation]
    bgs = nltk.bigrams(words_after_screening)
    fdist = nltk.FreqDist(bgs)
    return dict(fdist)

start_time = time.time()
# Generate a mapping dataframe
each_category_bigrams = {}

for category in brown_categories:
    each_category_bigrams[category] = {}
    category_words = brown.words(categories = category)
    for bigram in top_15_bigrams:
        bigrams_within_category = get_bigrams_frequency_by_word_list(category_words)
        try:
            each_category_bigrams[category][bigram] = bigrams_within_category[bigram]
        except Exception as e:
#             print("No such bigrams '{}' in {} category".format(e,category))
            each_category_bigrams[category][bigram] = 0
print("Finish!")
each_category_bigrams_df = pd.DataFrame(each_category_bigrams)
```

```python
plt.figure(figsize=(20,20))
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)


ax = sb.heatmap(each_category_bigrams_df, cmap = "Blues",annot=True, annot_kws={"size": 20})
ax.set_title("Heatmap of bigrams frequencies")
ax.title.set_size(25)
ax.set_ylabel('bigrams',fontdict={'fontsize':24})
ax.set_xlabel('category_name',fontdict={'fontsize':24})
plt.tight_layout()
plt.show()
```
